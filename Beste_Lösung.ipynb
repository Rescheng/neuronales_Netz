```python
import numpy as np

import pandas as pd
import math

import openpyxl
import time

start_time = time.perf_counter()

# hier kommt dein Programmcode




excel_file = 'Kontrolldatei_Instanz1.xlsx'

# Daten aus der Excel-Datei in ein pandas DataFrame laden
data = pd.read_excel(excel_file)

# Liste initialisieren, um die Daten zu speichern
Alle_Auftraege_insgesammt= []

episode_length = 10

# Schleife, um die Zeilen der Excel-Datei zu durchlaufen
for i, row in data.iterrows():
    # Wenn die Anzahl der Zeilen in der Liste episode_length erreicht hat, eine neue Liste hinzufügen
    if i % episode_length == 0:
        Alle_Auftraege_insgesammt.append([])
    # Daten aus der aktuellen Zeile hinzufügen
    Alle_Auftraege_insgesammt[-1].append(row.tolist())

Anzahl_der_Epochen = len(Alle_Auftraege_insgesammt)
#print(Alle_Auftraege_insgesammt)


def generate_lists(L):
    # Alle möglichen Kombinationen von Nullen und Einsen erzeugen
    for i in range(2):
        for j in range(2):
            for k in range(2):
                for l in range(2):
                    for m in range(2):
                        for n in range(2):
                            for o in range(2):
                                for p in range(2):
                                    for q in range(2):
                                        for r in range(2):
                                            # Liste mit den 10 Zahlen erzeugen
                                            lst = [i, j, k, l, m, n, o, p, q, r]
                                            # Liste zur Ergebnisliste hinzufügen
                                            L.append(lst)
                                            
                                                                                       
                                            
Action_Liste = []
generate_lists(Action_Liste)
#print(Action_Liste)
print('Länge Liste Action: ',len(Action_Liste))
#print(Action_Liste)

bestaction = []
z = 0
ungueltig = 0
Liste_best_action = []
Liste_best_score = []
print('Länge Anzahl Aufträge',len(Alle_Auftraege_insgesammt))

# jedes j ist eine Auftragsbündelung
#((1,9, 0,1
# 2,20, 47, 80
# .....))
print(Alle_Auftraege_insgesammt[0][0][1])

for j in range(len(Alle_Auftraege_insgesammt)-1):
    
    score = 0
    bestscore = 0
    Alter_Punkt = (0,0)
    aktueller_Punkt = (0,0)
    Ursprung = (0,0)
    
    bestaction = []
    
    
    score_Liste = []    
    z = 0
    for n in range(len(Action_Liste)):
        
        ungueltig = 0
        score = 0
        z += 1 
        Verbleibende_Kapa = 100
        Alter_Punkt = (0,0)
        aktueller_Punkt = (0,0)
        Ursprung = (0,0)
        belohnung = 0
        Kosten = 0 
        # an welchen exakten auftrag sind wir
        for i in range(10):


            action = Action_Liste[n][i]
            
            if action == 1:
                
                belohnung = Alle_Auftraege_insgesammt[j][i][1]
                aktueller_Punkt = (Alle_Auftraege_insgesammt[j][i][2],Alle_Auftraege_insgesammt[j][i][3])
                Kosten =  math.dist(Alter_Punkt,aktueller_Punkt)
                
                
                score = score + (belohnung - Kosten)
                Alter_Punkt = aktueller_Punkt
                Verbleibende_Kapa = Verbleibende_Kapa - Alle_Auftraege_insgesammt[j][i][0]


            if i == 9:
                #print(i)
                score = score - math.dist(Ursprung,aktueller_Punkt)
                #score -= math.dist(Ursprung,aktueller_Punkt)

            if Verbleibende_Kapa < 0:
                ungueltig = 1 
                score = 0
                
        

        score_Liste.append(score)

            
        if score > bestscore and ungueltig == 0 :
            bestscore = score
            bestaction = Action_Liste[n]
            
     
            
    #print('score:', max(score_Liste))
         
    
        #print(score)
    Liste_best_action.append(bestaction)
    Liste_best_score.append(bestscore)
    

print('Ergebnis: ',Liste_best_score[17])
# Neue Excel-Datei erstellen
wb = openpyxl.Workbook()

# Arbeitsblatt auswählen
ws = wb.active

# Überschriften für die Spalten setzen
ws['A1'] = 'Score'
for col in range(2, 12):
    ws.cell(row=1, column=col).value = f'Action {col-1}'

# Daten in die Excel-Tabelle schreiben
for row in range(2, 102):
    # Score-Wert in die erste Spalte schreiben
    ws.cell(row=row, column=1).value = Liste_best_score[row-2]
    
    # Action-Werte in die restlichen Spalten schreiben
    for col in range(2, 12):
        ws.cell(row=row, column=col).value = Liste_best_action[row-2][col-2]

# Excel-Datei speichern
wb.save('optimale_Loesung_100_Epochen.xlsx')

end_time = time.perf_counter()
elapsed_time = end_time - start_time

print("Laufzeit des Programms: {} Sekunden".format(elapsed_time))
```
